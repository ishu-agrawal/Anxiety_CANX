```python
data_keys = {'Depression': [3, 5, 10, 13, 16, 17, 21, 24, 26, 31, 34, 37, 38, 42],
             'Anxiety': [2, 4, 7, 9, 15, 19, 20, 23, 25, 28, 30, 36, 40, 41],
             'Stress': [1, 6, 8, 11, 12, 14, 18, 22, 27, 29, 32, 33, 35, 39]}


data_scores = {'Depression': [(0, 10), (10, 14), (14, 21), (21, 28)],
             'Anxiety': [(0, 8), (8, 10), (10, 15), (15, 20)],
             'Stress': [(0, 15), (15, 19), (19, 26), (26, 34)]}
```

```python
df_conditions = {'Anxiety_A':[],'Anxiety_I':[],'Anxiety_E':[],
                 'Depression_A':[],'Depression_I':[],'Depression_E':[],
                 'Stress_A':[],'Stress_I':[],'Stress_E':[]}
for i in data_keys["Anxiety"]:
    df_conditions['Anxiety_A'].append('Q' + str(i) + 'A')
    df_conditions['Anxiety_I'].append('Q' + str(i) + 'I')
    df_conditions['Anxiety_E'].append('Q' + str(i) + 'E')

for i in data_keys["Depression"]:
    df_conditions['Depression_A'].append('Q' + str(i) + 'A')
    df_conditions['Depression_I'].append('Q' + str(i) + 'I')
    df_conditions['Depression_E'].append('Q' + str(i) + 'E')

for i in data_keys["Stress"]:
    df_conditions['Stress_A'].append('Q' + str(i) + 'A')
    df_conditions['Stress_I'].append('Q' + str(i) + 'I')
    df_conditions['Stress_E'].append('Q' + str(i) + 'E')
```

```python
age_group = ['Below 20','20-24', '25-29', '30-34', '35-39', '40-49', '50-59', 'Above 60']
def age_grouping(data, string):
    age_group_func = [
    (data[string]<20),
     ((data[string]>=20) & (data[string]<25)),
     ((data[string]>=25) & (data[string]<30)),
     ((data[string]>=30) & (data[string]<35)),
     ((data[string]>=35) & (data[string]<40)),
     ((data[string]>=40) & (data[string]<50)),
     ((data[string]>=50) & (data[string]<60)),
     (data[string]>=60)
    ]
    data.insert(data.columns.get_loc('age'), "age_group" ,np.select(age_group_func, age_group))
    return data
    
data = age_grouping(data, 'age')
data.head(5)
```

```python
def Score(data, string):
    conditions = [
    ((data >= data_scores[string][0][0]) & (data < data_scores[string][0][1])),
    ((data >= data_scores[string][1][0]) & (data < data_scores[string][1][1])),
    ((data >= data_scores[string][2][0]) & (data < data_scores[string][2][1])),
    ((data >= data_scores[string][3][0]) & (data < data_scores[string][3][1])),
    (((data >= data_scores[string][3][1])))
    ]
    values = ['Normal','Mild', 'Moderate', 'Severe', 'Extremely Severe']
    df2 = pd.DataFrame({'score':data})
    df2['Score_Label'] = np.select(conditions, values)
    return df2
    
depr_target = Score(depr_scores, 'Depression')
depr_target.head()
```

```python
df_train = df_train.join(depr_target, how='left') 
df_train.shape
```

### MODEL BUILDING

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_train, depr_target, test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from category_encoders import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

pipe = make_pipeline(
    OneHotEncoder(use_cat_names=True),
    StandardScaler(),
    RandomForestClassifier(min_samples_leaf=10, n_jobs=-1, random_state=10, oob_score=True)
)

pipe.fit(X_train, y_train)
print('Train score:', pipe.score(X_train, y_train))
print('Test score:', pipe.score(X_test, y_test))
```

```python
rf = pipe.named_steps['randomforestclassifier']
colnames = pipe.named_steps['onehotencoder'].get_feature_names()
importances = pd.Series(rf.feature_importances_, colnames)

n = 10
plt.figure(figsize=(10,n/4))
plt.title(f'Top {n} features with onehotencoder')
importances.sort_values()[-n:].plot.barh();
```
